```python
import mysql.connector as sql   # used to establish a connection to the MySQL database and make queries.
#used for mathematical operations
import numpy as np 
#Seaborn simplifies the process of creating beautiful and informative graphs like scatter plots, bar charts, box plots, heatmaps, and more.
import seaborn as sns 
#Without this, the queries will be present in list or tuples. And to make it more readable, pandas can turn the data into structured table. 
#This is essential for working with data as DataFrames, which simplifies many data analysis tasks.
import pandas as pd
#this is essential for creating visualizations
import matplotlib.pyplot as plt

mydb= sql.connect(
    host= "localhost",
    user="root",
    password="1234",    
    database="catalonia_energy")
```

```python
energy_demand = pd.read_sql("SELECT * FROM energy_demand_in_catalunya", mydb)
energy_demand
```

```python
print(energy_demand.info())
```

Most of the data is numerical data

```python
energy_demand.shape
```

```python
energy_demand['DATA'] = pd.to_datetime(energy_demand['DATA'])

energy_demand['ANY'] = energy_demand['DATA'].dt.year

energy_demand
```

The table shows the hourly consumption of energy here in catalunya from 2005 until 2024. 

```python
plt.figure(figsize=(15, 8))
sns.barplot(data=energy_demand, x='ANY', y= 'TOTAL', color= 'navy')
plt.show()
```

```python


# Set a Seaborn style for the plot
sns.set(style="whitegrid")

# Calculate hourly averages
hourly_averages = energy_demand.iloc[:, 2:26].mean()

# Create the plot with Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_averages, marker='o', color='b')

# Add titles and labels
plt.title("Average Hourly Energy Demand")
plt.xlabel("Hour of the Day")
plt.ylabel("Energy Demand (units)")
plt.xticks(rotation=45)

plt.show()

```
